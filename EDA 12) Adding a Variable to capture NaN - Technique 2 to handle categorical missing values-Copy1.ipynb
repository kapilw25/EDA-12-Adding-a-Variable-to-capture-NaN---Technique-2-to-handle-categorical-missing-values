# Technique 2) Adding a Variable to capture Nan

```python
import pandas as pd
df = pd.read_csv('House Prices_train.csv')
# Download Link: https://www.kaggle.com/c/house-prices-advanced-regression-techniques/data?select=train.csv
df
```

```python
#cant read all 81 columns in above dataframe. So,
df.columns
```

```python
# lets focus on 4 particular columns / features
df = pd.read_csv('House Prices_train.csv',usecols=['BsmtQual', 'FireplaceQu', 'GarageType', 'SalePrice'])
df
# 'BsmtQual', 'FireplaceQu', 'GarageType' are Categorical features
# 'SalePrice' is dependent feature
```

```python
# creating a new feature to identify wehther the feature has Nan value or not
import numpy as np
```

```python
# adding a variable wrt features
df['BsmtQual_Var']= np.where(df['BsmtQual'].isnull(),1,0) # place 1 in new column if there is Nan value in Old Column
df['FireplaceQu_Var']= np.where(df['FireplaceQu'].isnull(),1,0)
df.head(25)
```

```python
df['BsmtQual'].mode() #finding the most frequent value in the column 'BsmtQual'
```

```python
frequent1 = df['BsmtQual'].mode()[0]  # defining the variable "frequent"
frequent1
```

```python
frequent2 = df['FireplaceQu'].mode()[0]  # defining the variable "frequent"
frequent2
```

```python
# place "frequent" value in new column if there is Nan value in Old Column
df['BsmtQual_Add']= df['BsmtQual'].fillna(frequent1)
df['FireplaceQu_Add']= df['FireplaceQu'].fillna(frequent2)
df.head(25)
```

```python
# re-place Nan value with "frequent" value in the OLD column
df['BsmtQual'].fillna(frequent1,inplace=True) # permanenet replacement by "inplace=True" command
df['FireplaceQu'].fillna(frequent2,inplace=True)
df.head(25) #no NaN values in orginal columns anymore
```

## if frequent categories are multiple, then NaN is replaced with New category
